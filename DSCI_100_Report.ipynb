```r
players_summary <- players |>
    select_if(is.numeric) |>
    pivot_longer(cols = everything(),
                 names_to = "Variable", 
                 values_to = "Value") |>
    group_by(Variable) |>
    summarize(Mean = round(mean(Value, na.rm = TRUE), 2),
              Median = round(median(Value, na.rm = TRUE), 2),
              Min = round(min(Value, na.rm = TRUE), 2),
              Max = round(max(Value, na.rm = TRUE), 2))
players_summary

sessions_summary <- sessions |>
    select_if(is.numeric) |>
    pivot_longer(cols = everything(),
                 names_to = "Variable", 
                 values_to = "Value") |>
    group_by(Variable) |>
    summarize(Mean = round(mean(Value, na.rm = TRUE), 2),
              Median = round(median(Value, na.rm = TRUE), 2),
              Min = round(min(Value, na.rm = TRUE), 2),
              Max = round(max(Value, na.rm = TRUE), 2))
sessions_summary
```

Broad question: What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?

Specific question: Can average game session duration predict whether a player is more likely to subscribe to the game-related newletter in the sessions dataset?

The sessions data holds information regarding start and end time of game duration which accounts for breaks in game play. The goal is to use the average sessions data to differentiate between players with long and short game play and how this impacts subscription. By merging these variables with the newsletter subscription data available in the players data set the broad and specific questions can be effectively targeted.

```r
sessions_min <- sessions |>
    mutate(sessions_duration = (original_end_time - original_start_time) / 1000 / 60)

avg_sessions_duration <- sessions_min |>
    filter(!is.na(sessions_duration)) |>
    group_by(hashedEmail) |>
    summarize(mean_sessions_duration = mean(sessions_duration), .groups = "drop")

merged_data <- players |>
    inner_join(avg_sessions_duration, by = "hashedEmail") |>
    select(-hashedEmail, -name)

head(merged_data)
```

```r
sessions_plot <- ggplot(merged_data, aes(x = subscribe, y = mean_sessions_duration, fill = subscribe)) +
    geom_col() +
    labs(title = "Average Session Duration by Newsletter Subscription",
         x = "Subscribed",
         y = "Average Session Duration (mins)") +
    theme(plot.title = element_text(size = 15))
sessions_plot

sessions_histogram <- ggplot(merged_data, aes(x = mean_sessions_duration, fill = subscribe)) +
    geom_histogram(binwidth = 10, alpha = 0.7, position = "identity") +
    labs(title = "Distribution of Average Session Duration by Newsletter Subscription",
         x = "Average Session Duration (mins)",
         y = "Number of Players") +
    facet_wrap(~subscribe) +
    theme(plot.title = element_text(size = 15))
sessions_histogram
```

The bar graph indicates that people who subscribe to the newsletter typically have longer average sessions compared to non-subscribers.

The histogram displays the distribution of session length and reveals where average session duration is similar and different between both groups. To make this pattern more clear it could be helpful to remove average game duration of 0 minutes. 

An ideal method for predicting the questions of interest is linear regression. This model is appropriate because the data deals with continuous predictor variables (mean_session_duration) that affect the response variable (subscribe). In the context of this model, I am assuming that subscription is determined independent of other players and that played_hours and mean_sessions_duration have little effect on each other (low multicollinearity). Similarly, this model is limited to linear relationships, therefore if the data exhibits any curves the model may have trouble with accuracy. Also, logical statements will have to be treated as numeric values. Separately, to process this data so linear regression can be applied I will filter out the 0 minute game sessions to concentrate on patterns in the data. Additionally, I plan to split the data into 70% training and 30% testing then carry out a 5-fold cross validation to prevent overfitting and assess the models performance on the unseen data. The cross validation will be used in lieu of a validation set.
