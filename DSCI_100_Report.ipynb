```r
library(tidyverse)
```

```r
players <- read_csv("players.csv")
sessions <- read_csv("sessions.csv")
head(players)
summary(players)
dim(players)
head(sessions)
summary(sessions)
dim(sessions)
```

The players dataset contains 196 rows and 7 columns which is the equivalent of 196 observations and 7 variables. hashedEmail and name variables can be ignored as they don’t provide any information that can be predicted upon. In contrast, demographic and behavioral data allows for the categorization of the likelihood an individual will subscribe. The variables that share the strongest positive relationship to subscription probability will be favoured in answering the broad and specific questions.

Variable Description
experience (chr) - Player experience level (beginner/regular/veteran)
subscribe (lgl) - Whether the player is subscribed to the newsletter
hashedEmail (chr) - Player username
played_hours (dbl) - Total hours played
name (chr) - Display name
gender (chr) - gender identity
age - Player age (years)

The sessions dataset includes 1535 rows and 5 columns which translates to 1535 observations and 5 variables. An issue presented in this table is poor readability of the units of time which will need to be converted for further data analysis. Also, start/end time and original start/end time appear redundant

Variable Description
hashedEmail (chr) - Player username
start_time (chr) - Human-readable start timestamp
end_time (chr) - Human-readable end timestamp
original_start_time (dbl) - Computer readable start timestamp
original_end_time (dbl) Computer readable end timestamp

```r
players_summary <- players |>
    select_if(is.numeric) |>
    pivot_longer(cols = everything(),
                 names_to = "Variable", 
                 values_to = "Value") |>
    group_by(Variable) |>
    summarize(Mean = round(mean(Value, na.rm = TRUE), 2),
              Median = round(median(Value, na.rm = TRUE), 2),
              Min = round(min(Value, na.rm = TRUE), 2),
              Max = round(max(Value, na.rm = TRUE), 2))
players_summary

sessions_summary <- sessions |>
    select_if(is.numeric) |>
    pivot_longer(cols = everything(),
                 names_to = "Variable", 
                 values_to = "Value") |>
    group_by(Variable) |>
    summarize(Mean = round(mean(Value, na.rm = TRUE), 2),
              Median = round(median(Value, na.rm = TRUE), 2),
              Min = round(min(Value, na.rm = TRUE), 2),
              Max = round(max(Value, na.rm = TRUE), 2))
sessions_summary
```
